```python
import pandas as pd
films = pd.read_csv('C:\\python\\pyda-22\\pandas2\\ratings.csv')

def clas(row):
    if row.rating.mean() <= 2:
        row['class'] = 'Низкий рейтинг'
    elif row.rating.mean() <= 4:
        row['class'] = 'Средний рейтинг'
    else:
        row['class'] = 'Высокий рейтинг'
    return row
    
films.groupby('movieId').apply(clas).sort_values(['movieId']).head()
```

```python
keywords =  pd.read_csv('C:\\python\\pyda-22\\pandas2\\keywords.csv')

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

geo_data_1 = {}

for key in geo_data:
    for city in geo_data[key]:
        geo_data_1[city] = key

def geo(row):
    for key in geo_data_1:
        if key in row.keyword:
            row['region'] = geo_data_1[key]
    return row
keywords.apply(geo, axis=1).fillna('undefined').head()
```
