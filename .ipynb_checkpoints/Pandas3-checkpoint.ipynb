```python
import pandas as pd
import numpy as np
data = pd.read_csv('C:\\python\\pyda-22\\visit_log.csv', sep=';')
data.loc[(data.traffic_source == 'yandex') | (data.traffic_source == 'google'), 'source_type'] = 'organic'
data.loc[((data.traffic_source == 'paid') 
        | (data.traffic_source == 'email')) 
        & (data.region == 'Russia'), 'source_type'] = 'ad'
data.loc[((data.traffic_source == 'paid') 
        | (data.traffic_source == 'email')) 
        & (data.region != 'Russia'), 'source_type'] = 'other'

data['source_type'] = data.apply(lambda row: row.traffic_source if pd.isnull(row.source_type) else row.source_type, axis=1)
data.head()
```

```python
rating = pd.read_csv('C:\\python\\pyda-22\\ml-latest-small\\ratings.csv')
rating.head()
rating = (rating.groupby(['userId'])
.agg({'movieId':['count'], 'timestamp':['max', 'min']})
)

rating = (rating.loc[rating['movieId']['count'] > 100]
          .apply(lambda row: row['timestamp']['max'] - row['timestamp']['min'], axis=1))
print(f'среднее время жизни = {round(sum(rating)/len(rating)/3600/24)} дня')
```
