```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python


list_=['python','парсинг']

def search_habr(words):
    kom_news = pd.DataFrame()
    #50 страниц доступно на хабр 
    for i in range(1, 51):
        req = requests.get('https://habr.com/ru/all/page'+str(i))
        time.sleep(0.3)
        soup = BeautifulSoup(req.text, 'html.parser')
        posts = soup.find_all('article', class_='tm-articles-list__item')
        #проходимся по постам в рамках одной страницы
        for post in posts:
            if post.find('div', class_='article-formatted-body article-formatted-body_version-2'):
                preview = post.find('div', class_='article-formatted-body article-formatted-body_version-2').text
            else:
                preview = post.find('div', class_='article-formatted-body article-formatted-body_version-1').text
            if post.find('a', class_='tm-article-snippet__title-link'):
                title = post.find('a', class_='tm-article-snippet__title-link').text
            else:
                title = post.find('h2', class_='tm-megapost-snippet__title').text
            date = post.find('time').get('datetime')
            if post.find('a', class_='tm-article-snippet__title-link'):
                href = post.find('a', class_='tm-article-snippet__title-link').get('href')
            else:
                href = post.find('a', class_='tm-megapost-snippet__link tm-megapost-snippet__card').get('href')
            #проходимся по ключевым словам в поиске                 
            for word in words:
                if (word in preview) or (word in title):
                    date = pd.to_datetime(date, dayfirst=True).date()
                    href = 'https://habr.com'+href
                    row = {'date': date, 'title': title, 'href': href}
                    kom_news = pd.concat([kom_news, pd.DataFrame([row])])  
    kom_news.reset_index()
    return kom_news

search_habr(list_)
```

```python
EMAIL = ['xxx@x.ru', 'yyy@y.com']

def email_check(list_):
    breaches = pd.DataFrame()

    URL = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
    headers = {
        'Accept': 'application/json, text/plain, */*',
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
        'Connection': 'keep-alive',
        'Content-Length': '34',
        'Content-Type': 'application/json;charset=UTF-8',
        'DNT': '1',
        'Host': 'identityprotection.avast.com',
        'Origin': 'https://www.avast.com',
        'Referer': 'https://www.avast.com/',
        'sec-ch-ua': '" Not;A Brand";v="99", "Google Chrome";v="91", "Chromium";v="91"',
        'sec-ch-ua-mobile': '?0',
        'Sec-Fetch-Dest': 'empty',
        'Sec-Fetch-Mode': 'cors',
        'Sec-Fetch-Site': 'same-site',
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Vaar-Header-App-Build-Version': '1.0.0',
        'Vaar-Header-App-Product': 'hackcheck-web-avast',
        'Vaar-Header-App-Product-Name': 'hackcheck-web-avast',
        'Vaar-Version': '0'
    }

    req = requests.post(URL, json = {"emailAddresses": list_}, headers = headers) 
    b = json.loads(req.text)

    for i in b['breaches']:
        site = b['breaches'][i]['site']
        description = b['breaches'][i]['description']
        date = pd.to_datetime(b['breaches'][i]['publishDate'], dayfirst=True).date()
        row = {'date': date, 'site': site, 'description': description}
        breaches = pd.concat([breaches, pd.DataFrame([row])])  
    return(breaches)

email_check(EMAIL)
```
